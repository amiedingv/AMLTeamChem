```python
import pandas as pd
import matplotlib.pyplot as plt
import collections
```

```python
training_data = pd.read_csv('./dataset/train.csv')
```

```python
training_data.head()
```

```python
sentences = training_data['SENTENCE'].tolist()

labels = {}
for sentence in sentences:
    smells = sentence.split(',')
    for smell in smells:
        if smell in labels:
            labels[smell] += 1
        else:
            labels[smell] = 1
```

```python
# smells and their count
labels = {k: v for k, v in sorted(labels.items(), key=lambda item: item[1], reverse=True)}
labels
```

```python
dates = list(labels.keys())           
prices = list(labels.values()) 
plt.plot(dates, prices, '-')
```

```python
# plot of different smells and their corresponding count - highly unbalanced
plt.bar(*zip(*labels.items()))
plt.show()
```

```python
# checking for duplicate values in smiles - no duplicates found
smiles = training_data['SMILES'].tolist()
print([item for item, count in collections.Counter(smiles).items() if count > 1])
```

```python
l10 = 0
l25 = 0
l50 = 0
l100 = 0
l200 = 0
l300 = 0

for smile in smiles:
    if(len(smile) < 10):
        l10 += 1
    elif(len(smile) < 25):
        l25 += 1
    elif(len(smile) < 50):
        l50 += 1
    elif(len(smile) < 100):
        l100 += 1
    elif(len(smile) < 200):
        l200 += 1
    elif(len(smile) < 300):
        l300 += 1
    else:
        print(smile)
    
```

```python
x = ['< 10', '10 - 25', '25 - 50', '50 - 100', '100 - 200', '200+']
energy = [l10, l25, l50, l100, l200, l300]

x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, energy, color='green')
plt.xlabel("Energy Source")
plt.ylabel("Energy Output (GJ)")
plt.title("")

plt.xticks(x_pos, x)

plt.show()
```
