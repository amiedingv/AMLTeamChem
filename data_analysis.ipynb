```python
import pandas as pd
import matplotlib.pyplot as plt
import collections
```

```python
training_data = pd.read_csv('./dataset/train.csv')
```

```python
training_data.head()
```

```python
sentences = training_data['SENTENCE'].tolist()

labels = {}
for sentence in sentences:
    smells = sentence.split(',')
    for smell in smells:
        if smell in labels:
            labels[smell] += 1
        else:
            labels[smell] = 1
```

```python
# smells and their count
labels = {k: v for k, v in sorted(labels.items(), key=lambda item: item[1], reverse=True)}
labels
```

```python
# plot of different smells and their corresponding count - highly unbalanced
plt.bar(*zip(*labels.items()))
plt.show()
```

```python
# checking for duplicate values in smiles - no duplicates found
smiles = training_data['SMILES'].tolist()
print([item for item, count in collections.Counter(smiles).items() if count > 1])
```

```python
# checking for repeated values in sentence
sent = training_data['SENTENCE'].tolist()
print(len([item for item, count in collections.Counter(sent).items() if count > 1]))
```

```python
labels = dict(collections.Counter(sent))
labels = {k: v for k, v in sorted(labels.items(), key=lambda item: item[1], reverse=True)}
labels
```
